```python
import pandas as pd
```

```python
df = pd.read_csv('1.csv',encoding='gbk', sep=',', header=None)
# df
```

```python
df = df.fillna('nnnnaaaannnn')
# df
```

```python
alllist = []
for i in df:
    alllist.extend(df[i].to_list())
#     print(df[i].to_list())
alllist
```

```python
alllist.remove('nnnnaaaannnn')
```

```python
for item in alllist.copy():
    if item == 'nnnnaaaannnn':
        alllist.remove(item)
    else:
        pass
```

```python
alllist
```

```python
import pypinyin

# 不带声调的(style=pypinyin.NORMAL)
def pinyin(word):
    s = ''
    for i in pypinyin.pinyin(word, style=pypinyin.NORMAL):
        s += ''.join(i)
    return s
```

```python
names = []
for name in alllist:
    names.append(pinyin(name))
#     print(pinyin(name))
```
